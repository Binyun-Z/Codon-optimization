```python
import pandas as pd
```

```python
CLIB = pd.read_csv('./cerevisiae.csv')
```

```python
CLIBgroup = CLIB.groupby('AmOneLet')
list(CLIBgroup)
```

```python
len(list(CLIBgroup))
```

```python
for aa,group in CLIBgroup:
    group['Percentage'] = group['frequency']/group['frequency'].sum()
```

```python
CLIB['Percentage'] = CLIB['frequency'] / CLIB.groupby('AmOneLet')['frequency'].transform('sum')
CLIB
```

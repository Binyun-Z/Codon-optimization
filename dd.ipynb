```python
import pandas as pd
```

```python
rscu0 = pd.read_csv('./cerevisiae_gfp_200_RSCUmin_0.csv')
rscu05 = pd.read_csv('./cerevisiae_gfp_200_RSCUmin_05.csv')
rscu08 = pd.read_csv('./cerevisiae_gfp_200_RSCUmin_08.csv')
```

```python
# 创建区间
bins = [x / 100 + 0.6 for x in range(31)]  # [0.0, 0.1, 0.2, ..., 1.0]

# 使用 cut 函数将小数分配到对应的区间
rscu0['bin'] = pd.cut(rscu0['CAI'], bins)

# 使用 value_counts 函数统计每个区间的小数数量
counts0 =rscu0['bin'].value_counts().sort_index()

print(counts0)
```

```python


# 使用 cut 函数将小数分配到对应的区间
rscu05['bin'] = pd.cut(rscu05['CAI'], bins)

# 使用 value_counts 函数统计每个区间的小数数量
counts05 =rscu05['bin'].value_counts().sort_index()

print(counts05)
```

```python


# 使用 cut 函数将小数分配到对应的区间
rscu08['bin'] = pd.cut(rscu08['CAI'], bins)

# 使用 value_counts 函数统计每个区间的小数数量
counts08 =rscu08['bin'].value_counts().sort_index()

print(counts08)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.6, 0.89, 0.01)  # 生成从0.6到0.9的小数，间隔为0.01
y = counts0.values

plt.bar(x,y,width=0.008,align='edge',label = 'aaa')
# 添加图例
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import  numpy as np
w = 0.008
c = "#4CAF50"
x = np.arange(0.6, 0.89, 0.01)  # 生成从0.6到0.9的小数，间隔为0.01
# 创建一个 3x1 的子图布局，并共享横坐标
fig, axes = plt.subplots(3, 1, figsize=(4, 10), sharex=True)

# 绘制第一个子图
axes[0].bar(x, counts0.values,width=w,align='edge',color = '#4CAF50',label='RSCU 0')
axes[0].text(0.8, 0.9, 'RSCU_min 0',transform=axes[0].transAxes, fontsize=12, ha='center', va='center')
# axes[0].legend()

# 绘制第二个子图
axes[1].bar(x, counts05.values,width=w,align='edge',color = 'g',label='RSCU 0.5')
axes[1].text(0.8, 0.9, 'RSCU_min 0.5',transform=axes[1].transAxes, fontsize=12, ha='center', va='center')
# axes[1].legend()

# 绘制第三个子图
axes[2].bar(x, counts08.values,width=w,align='edge',color = '#556B2F',label='RSCU 0.8')
axes[2].text(0.25, 0.9, 'RSCU_min 0.8',transform=axes[2].transAxes, fontsize=12, ha='center', va='center')
# axes[2].legend()
plt.xlabel('CAI',fontsize=15)
fig.text(0.02, 0.5, 'Count', va='center',fontsize=15, rotation='vertical')

# 调整子图间的距离为0
plt.subplots_adjust(hspace=0)
plt.show()
plt.savefig('RSCU.png', dpi=600)
```
